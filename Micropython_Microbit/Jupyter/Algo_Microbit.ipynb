# Algorigrammes pour Microbit

```python
import schemdraw
from schemdraw import flow
schemdraw.__version__
```

```python
def logigramme_si(sch,drapeau) :
    A=sch.add(flow.LINE,d='down',l=1)
    sch.add(flow.ARROWHEAD,d='down')
    sch.add(flow.LINE,d='down',l=0.05)
    sch.add(flow.Decision(w=4, h=2, E='OUI', S='NON',label=drapeau[1]))
    sch.add(flow.LINE,d='down',l=0.75)
    if drapeau[-1] : sch.add(flow.Box(w=3, h=2,label=drapeau[3]))#pour différencier si de si... sinon
    else : sch.add(flow.LINE,d='down',l=2)
    sch.add(flow.LINE,d='down',l=1)
    sch.add(flow.LINE,d='right', xy=[A.end[0]+2,A.end[1]-1.05],l=1.5)
    sch.add(flow.LINE,d='down',l=0.75)
    sch.add(flow.Box(w=3, h=2,label=drapeau[2]))
    sch.add(flow.LINE,d='down',l=1.5)
    sch.add(flow.LINE,d='left',l=3.45)
    sch.add(flow.ARROWHEAD,d='left')
    sch.add(flow.LINE,d='left',l=0.05)    
    sch.add(flow.LINE,d='down',l=1)
    sch.add(flow.ARROWHEAD,d='down')
```

```python
def Boucle_haut(sch,nom) :
    A=sch.add(flow.LINE,d='right',l=1)
    sch.add(flow.LINE,to=[A.end[0]+.5,A.end[1]-.5])
    sch.add(flow.LINE,d='down',l=1)
    B=sch.add(flow.LINE,d='left',l=3)
    C=sch.add(flow.LINE,d='up',l=1)
    sch.add(flow.LINE,d='right',xy=[C.end[0]+0.75,C.end[1]-0.25],l=0,rgtlabel=nom)
    sch.add(flow.LINE, xy=[C.end[0],C.end[1]],to=[C.end[0]+0.5,C.end[1]+.5])
    sch.add(flow.LINE,d='right',l=1)
    sch.add(flow.LINE,d='down', xy=[B.end[0]+1.5,B.end[1]],l=1)
```

```python
def Boucle_bas(sch,nom) :
    sch.add(flow.LINE,d='right',l=1.5)
    E=sch.add(flow.LINE,d='down',l=1.0)
    sch.add(flow.LINE, to=[E.end[0]-0.5,E.end[1]-0.5])
    F=sch.add(flow.LINE,d='left',l=2)
    sch.add(flow.LINE,to=[F.end[0]-0.5,F.end[1]+0.5])
    sch.add(flow.LINE,d='up',l=1.0)
    sch.add(flow.LINE,d='right',l=1.5)
    sch.add(flow.LINE,d='right',xy=[F.end[0],F.end[1]+0.75],l=0,rgtlabel=nom)
```

```python
def fleche(sch,taille=1) :
    sch.add(flow.LINE,d='down', l=taille)
    sch.add(flow.ARROWHEAD,d='down')
    sch.add(flow.LINE,d='down',l=0.05)
```

```python
e = schemdraw.Drawing(fontsize=14.0)
texte1=['DÉBUT','Bouton\n appuyé ?','Allumer LED\n centrale','']
depart=e.add(flow.Start(w=2, h=1,label=texte1[0]))
fleche(e)
Boucle_haut(e,'Boucle')
logigramme_si(e,texte1)
Boucle_bas(e,'Sans fin')
e.draw()
#e.save('../../Images/algo_si_µB.svg')
```

```python
e = schemdraw.Drawing(fontsize=14.0)
texte1=['DÉBUT','Bouton\n appuyé ?','Allumer LED\n centrale','Éteindre LED\n centrale']
depart=e.add(flow.Start(w=2, h=1,label=texte1[0]))
fleche(e)
Boucle_haut(e,'Boucle')
logigramme_si(e,texte1)
Boucle_bas(e,'Sans fin')
e.draw()
#e.save('../../Images/algo_si_sinon_µB.svg')
```

```python
e = schemdraw.Drawing(fontsize=14.0)
texte1=['DÉBUT','Bouton B\n appuyé ?','Allumer LED\n centrale','']
texte2=['DÉBUT','Bouton A\n appuyé ?','Éteindre LED\n centrale','']
depart=e.add(flow.Start(w=2, h=1,label=texte1[0]))
fleche(e)
Boucle_haut(e,'Boucle')
logigramme_si(e,texte1)
logigramme_si(e,texte2)
Boucle_bas(e,'Sans fin')
e.draw()
#e.save('../../Images/algo_si2_µB.svg')
```

```python
e = schemdraw.Drawing(fontsize=10.0)
texte1=['DÉBUT','Nb de fois BP \n appuyé impair ?','Inverser\n état lampe','']
depart=e.add(flow.Start(w=2, h=1,label=texte1[0]))
fleche(e)
e.add(flow.Box(w=3, h=1,label='Éteindre\n Lampe'))
fleche(e)
Boucle_haut(e,'Boucle')
logigramme_si(e,texte1)
Boucle_bas(e,'Sans fin')
e.draw()
# e.save('../../Images/algo_telemetre_µB.svg')
```
